## Ejemplo de análisis de datos en Psicolingüística

```python
import pandas as pd
import pingouin as pg
import seaborn as sns
```

## Cargamos datos

```python


data = pd.read_spss("16-30_Produccion.sav")
df = data[["Cod", "Edad", "Sexo", "Total", "Edad_Cat", 
               "Nombre", "Adjetivos", "Verbos", "Adverbios", 
               "Auxiliares", "Preposicion"]]
df
```

## Exploramos posibles asociaciones entre variables

```python
sns.scatterplot(x="Edad", y="Total", data=df)
```

```python
sns.boxplot(x="Edad_Cat", y="Total", data=df)
```

```python
sns.boxplot(x="Sexo", y="Total", data=df)
```

```python
sns.pairplot(data=df[["Edad","Nombre", "Adjetivos", "Verbos", "Adverbios", 
               "Auxiliares", "Preposicion"]])
```

## Análisis de correlación

```python
pg.pairwise_corr(columns=["Edad","Nombre", "Adjetivos", "Verbos", "Adverbios", 
               "Auxiliares", "Preposicion"], data=df, method="kendall", padjust="bonf").round(3)
```
