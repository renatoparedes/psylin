## Cargar datos

```python
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/gastonstat/matrix4sl/master/data/starwars.csv"
)
df
```

## Gr√°fico de cajas

```python
sns.boxplot(x="gender", y="weight", data=df)
```

```python
df.groupby("gender")["weight"].describe()
```

```python
sns.boxplot(x="jedi", y="weight", data=df)
```

```python
df.groupby("jedi")["weight"].describe()
```

## Contraste de muestras

### Param√©trico

```python
pg.pairwise_tests(between="gender", dv="weight", data=df, effsize="cohen", parametric=True)
```

## No param√©trico

```python
pg.normality(dv="weight", group="jedi", data=df)
```

```python
pg.pairwise_tests(between="jedi", dv="weight", data=df, effsize="CLES", parametric=False)
```
